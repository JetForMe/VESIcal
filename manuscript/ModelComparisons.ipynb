## 4.1 Model Comparisons

### We can use one sample composition to show comparisons between model calcs, user can change this sample to whatever they want! This becomes a useful tool for anyone to automatically compare model results for all calcs for any sample of interest.

# Dissolved Volatiles

```python
# magmasat_vols = myfile.calculate_dissolved_volatiles(temperature=900.0, 
#                                 pressure=2000.0, X_fluid=0.5, verbose=False)

magmasat_basalts = basalts.calculate_dissolved_volatiles(temperature=900.0, 
                                pressure=2000.0, X_fluid=0.5, verbose=False)

magmasat_rhyolites = rhyolites.calculate_dissolved_volatiles(temperature=900.0, 
                                pressure=2000.0, X_fluid=0.5, verbose=False)

magmasat_alkaline = alkaline.calculate_dissolved_volatiles(temperature=900.0, 
                                pressure=2000.0, X_fluid=0.5, verbose=False)

# shish_vols = myfile.calculate_dissolved_volatiles(model='Shishkina', temperature=900.0,
#                                 pressure=2000.0, X_fluid=0.5)

# dix_vols = myfile.calculate_dissolved_volatiles(model='Dixon', temperature=900.0,
#                                           pressure=2000.0, X_fluid=0.5)

# iac_vols = myfile.calculate_dissolved_volatiles(model='IaconoMarziano', temperature=900.0,
#                                           pressure=2000.0, X_fluid=0.5)
```

```python
models = ['Shishkina', 'Dixon', 'IaconoMarziano']
fig, ax = plt.subplots(len(models), figsize=(8,24))

for i in range(len(models)):
    print(models[i])
    ax[i].set_ylabel('Dissolved H$_2$O wt% \n' + str(models[i]))
    ax[i].set_xlabel('Dissolved H$_2$O wt%, MagmaSat')
    result_basalts = basalts.calculate_dissolved_volatiles(model=models[i], temperature=1200.0,
                                pressure=2000.0, X_fluid=0.5)
    result_rhyolites = rhyolites.calculate_dissolved_volatiles(model=models[i], temperature=1200.0,
                                pressure=2000.0, X_fluid=0.5)
    result_alkaline = alkaline.calculate_dissolved_volatiles(model=models[i], temperature=1200.0,
                                pressure=2000.0, X_fluid=0.5)
    
    
    im = ax[i].scatter(magmasat_basalts["H2O_liq_VESIcal"], result_basalts["H2O_liq_VESIcal"], c=basalts.data['SiO2'], label="Basalts")
    ax[i].scatter(magmasat_rhyolites["H2O_liq_VESIcal"], result_rhyolites["H2O_liq_VESIcal"], color=v.color_list[1], label="Rhyolites")
    ax[i].scatter(magmasat_alkaline["H2O_liq_VESIcal"], result_alkaline["H2O_liq_VESIcal"], color=v.color_list[2], label="Alkaline")
    xy_1to1 = np.linspace(magmasat_alkaline["H2O_liq_VESIcal"].min()-0.5,magmasat_rhyolites["H2O_liq_VESIcal"].max()+0.5,100)
    ax[i].plot(xy_1to1, xy_1to1, color='k')
fig.colorbar(im, ax=ax[0])
ax[0].legend()
```
