## 4.1 Model Comparisons

### We can use one sample composition to show comparisons between model calcs, user can change this sample to whatever they want! This becomes a useful tool for anyone to automatically compare model results for all calcs for any sample of interest.

```python
import sys
sys.path.insert(0, '../../')

import VESIcal as v
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from IPython.display import display, HTML
%matplotlib inline
```

```python
basalts = v.ExcelFile("../../testDataSets/basalts.xlsx")
alkaline = v.ExcelFile("../../testDataSets/alkaline.xlsx")
rhyolites = v.ExcelFile("../../testDataSets/rhyolites.xlsx")

#MAGMASAT
magmasat_basalts = basalts.calculate_dissolved_volatiles(temperature=1000.0, 
                                pressure=2000.0, X_fluid=0.5, verbose=False)
magmasat_alkaline = alkaline.calculate_dissolved_volatiles(temperature=1000.0, 
                                pressure=2000.0, X_fluid=0.5, verbose=False)
magmasat_rhyolites = rhyolites.calculate_dissolved_volatiles(temperature=1000.0, 
                                pressure=2000.0, X_fluid=0.5, verbose=False)

#SHISHKINA
shishkina_basalts = basalts.calculate_dissolved_volatiles(temperature=1000.0, 
                                pressure=2000.0, X_fluid=0.5, model="Shishkina")
shishkina_alkaline = alkaline.calculate_dissolved_volatiles(temperature=1000.0, 
                                pressure=2000.0, X_fluid=0.5, model="Shishkina")
shishkina_rhyolites = rhyolites.calculate_dissolved_volatiles(temperature=1000.0, 
                                pressure=2000.0, X_fluid=0.5, model="Shishkina")

#DIXON
dixon_basalts = basalts.calculate_dissolved_volatiles(temperature=1000.0, 
                                pressure=2000.0, X_fluid=0.5, model="Dixon")
dixon_alkaline = alkaline.calculate_dissolved_volatiles(temperature=1000.0, 
                                pressure=2000.0, X_fluid=0.5, model="Dixon")
dixon_rhyolites = rhyolites.calculate_dissolved_volatiles(temperature=1000.0, 
                                pressure=2000.0, X_fluid=0.5, model="Dixon")

#IACONO-MARZIANO
IM_basalts = basalts.calculate_dissolved_volatiles(temperature=1000.0, 
                                pressure=2000.0, X_fluid=0.5, model="IaconoMarziano")
IM_alkaline = alkaline.calculate_dissolved_volatiles(temperature=1000.0, 
                                pressure=2000.0, X_fluid=0.5, model="IaconoMarziano")
IM_rhyolites = rhyolites.calculate_dissolved_volatiles(temperature=1000.0, 
                                pressure=2000.0, X_fluid=0.5, model="IaconoMarziano")
```

<hr>

```python
plt.figure(figsize=(12,8))
plt.xlabel("Dissolved H$_2$O wt%, MagmaSat")
plt.ylabel("Dissolved H$_2$O wt%, other models")

plt.scatter(magmasat_basalts["H2O_liq_VESIcal"],
            shishkina_basalts["H2O_liq_VESIcal"],
            label="Shishkina")
plt.scatter(magmasat_basalts["H2O_liq_VESIcal"],
            dixon_basalts["H2O_liq_VESIcal"],
            label="Dixon")
plt.scatter(magmasat_basalts["H2O_liq_VESIcal"],
            IM_basalts["H2O_liq_VESIcal"],
            label="Iacono-Marziano")
xy_1to1 = np.linspace(magmasat_basalts["H2O_liq_VESIcal"].min()-0.5,IM_basalts["H2O_liq_VESIcal"].max()+0.5,100)
plt.plot(xy_1to1, xy_1to1, label="1:1 Line", color='k')
plt.title("Basalts")
plt.legend()
plt.xlim(left=2.5, right=4)
plt.ylim(bottom=2.5, top=4)
```

```python
plt.figure(figsize=(12,8))
plt.xlabel("Dissolved H$_2$O wt%, MagmaSat")
plt.ylabel("Dissolved H$_2$O wt%, other models")

plt.scatter(magmasat_alkaline["H2O_liq_VESIcal"],
            shishkina_alkaline["H2O_liq_VESIcal"],
            label="Shishkina")
plt.scatter(magmasat_alkaline["H2O_liq_VESIcal"],
            dixon_alkaline["H2O_liq_VESIcal"],
            label="Dixon")
plt.scatter(magmasat_alkaline["H2O_liq_VESIcal"],
            IM_alkaline["H2O_liq_VESIcal"],
            label="Iacono-Marziano")
xy_1to1 = np.linspace(magmasat_alkaline["H2O_liq_VESIcal"].min()-0.5,IM_alkaline["H2O_liq_VESIcal"].max()+0.5,100)
plt.plot(xy_1to1, xy_1to1, label="1:1 Line", color='k')
plt.title("Alkaline")
plt.legend()
plt.xlim(left=2, right=4)
plt.ylim(bottom=2, top=4)
```

```python
plt.figure(figsize=(12,8))
plt.xlabel("Dissolved H$_2$O wt%, MagmaSat")
plt.ylabel("Dissolved H$_2$O wt%, other models")

plt.scatter(magmasat_rhyolites["H2O_liq_VESIcal"],
            shishkina_rhyolites["H2O_liq_VESIcal"],
            label="Shishkina")
plt.scatter(magmasat_rhyolites["H2O_liq_VESIcal"],
            dixon_rhyolites["H2O_liq_VESIcal"],
            label="Dixon")
plt.scatter(magmasat_rhyolites["H2O_liq_VESIcal"],
            IM_rhyolites["H2O_liq_VESIcal"],
            label="Iacono-Marziano")
xy_1to1 = np.linspace(magmasat_rhyolites["H2O_liq_VESIcal"].min()-5,IM_rhyolites["H2O_liq_VESIcal"].max()+5,100)
plt.plot(xy_1to1, xy_1to1, label="1:1 Line", color='k')
plt.title("Rhyolites")
plt.legend()
plt.xlim(left=1, right=3.7)
plt.ylim(bottom=1, top=3.7)
```
